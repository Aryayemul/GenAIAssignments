```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

###load model


```python
#load model
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda") #use gpu if available
```

### prompt

```python
prompt = "A futuristic cyberpunk city at sunset, neon lights, digital art, ultra detailed"
```

### Generate Image

```python
image = pipe(prompt).images[0] #generate image

image.save("creative_artwork.png")#save image

#display image
plt.imshow(image)
plt.axis("off")
plt.show()
```

### Improving artwork

```python
image = pipe(
    prompt,
    num_inference_steps=50,
    guidance_scale=7.5
).images[0]
```

### Generating multiple artworks

```python
for i in range(3):
    image = pipe(prompt).images[0]
    image.save(f"artwork_{i}.png")
```
